```python
from pandas_datareader import data as web
from datetime import datetime

start = datetime(2020, 6, 18)
end = datetime(2021, 6,18)

dfpharm = web.DataReader(['PFE', 'MRK', 'NVS', 'JNJ', 'BMI', 'GSK', 'ABT', 'AZN', 'SNY', 'LLY', 'NVS'],'yahoo',start=start,end=end)['Adj Close']
dfpharm
```

```python
tickers_list = ['PFE', 'MRK', 'NVS', 'JNJ', 'BMI', 'GSK', 'ABT', 'AZN', 'SNY', 'LLY', 'NVS']

# Import pandas
data = pd.DataFrame(columns=tickers_list)

# Fetch the data

for ticker in tickers_list:
    data[ticker] = yf.download(ticker,'2020-06-18','2021-06-18')['Adj Close']

# Print first 5 rows of the data
data
```

```python
import seaborn as sns
```

```python
sns.heatmap(data.corr(), annot = True)
# часто взлетают в определенные периоды времени одни и те же вещи, я не исключаю тупо совпадения, они явно есть и их много
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Plot all the close prices
((data.pct_change()+1).cumprod()).plot(figsize=(10, 7))

# Show the legend
plt.legend()

# Define the label for the title of the figure
plt.title("Adjusted Close Price", fontsize=16)

# Define the labels for x-axis and y-axis
plt.ylabel('Price', fontsize=14)
plt.xlabel('Year', fontsize=14)

# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
